**Barplot:**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your DataFrame and 'origin' is a categorical variable

plt.figure(figsize=(14, 8))  # Increase the figure size
sns.barplot(data=r, x='model_year', y='mpg', hue='origin', ci=None)

plt.xlabel('Model Year', fontsize=12)
plt.ylabel('MPG', fontsize=12)
plt.xticks(rotation=45, fontsize=10)  # Rotate x-axis labels for better readability
plt.yticks(fontsize=10)
plt.legend(title='Origin', fontsize=10, title_fontsize=12, loc='upper left')  # Adjust legend
plt.title('MPG by Model Year and Origin', fontsize=15)

plt.tight_layout()  # Adjust layout to fit everything nicely
plt.show()
```

```python
res = sns.barplot(x='origin',y='mpg',hue='model_year',hue_order=[82,78,80,76,72,70,74,72],data=r,color='yellow',order=['japan','europe','usa'])
plt.show()
```

```python
d=sns.barplot(x='cylinders',y='mpg',data=r,palette='Blues',orient='h')
plt.show()
```

```python

sh=sns.barplot(x='cylinders',y='mpg',data=r,saturation=10,errcolor='r',errwidth=10,capsize=0.2)
plt.show()
```

```python
sns.set(style='darkgrid')
sh=sns.barplot(x='cylinders',y='mpg',data=r,saturation=1000,alpha=0.5)
plt.show()
```

```python
ar=sns.scatterplot(x='horsepower',y='weight',hue='origin',data=r,palette='rocket',style='origin',size='origin',sizes=(50,100),alpha=0.5)
plt.show()
```

**CountPlot**

```python
ds=sns.countplot(x='model_year',data=r,palette='viridis')
plt.show()
```

```python
a=sns.countplot(y='origin',data=r,color='green')
plt.show()
```

```python
d = sns.countplot(x='origin', data=r, hue='model_year', palette='tab10')
plt.show()
```

**Lineplot:**

```python
sc = sns.lineplot(x='mpg',y='model_year',hue='origin',style='origin',data=r.head(100),markers=['o','>','s'])
plt.show()
```

```python
a=sns.lineplot(x='cylinders',y='mpg',data=r.head(120),hue='origin',palette='Accent',legend='brief')
plt.grid()
plt.title("Seaborn")
plt.show()
```

**Histograph:**

```python
asd=sns.displot(r['mpg'],bins=[10,15,20,25,30,35,40,45,50],kde=True,rug=True,color='b')
plt.show()
```

```python
aed=sns.displot(r['mpg'],kde=True,rug=False,color='b',log_scale=True)
plt.show()
```
